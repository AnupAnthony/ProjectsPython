```python
# dropping the category since it has many null values

df.drop(["Market Category"], axis=1, inplace=True)
```

```python
# Delete records that have null data

df = df.dropna()
```

```python
#dropping the duplicate values

df.drop_duplicates(inplace = True)
```

```python
df.hist(figsize=(12, 8),facecolor = 'lightgreen')
plt.tight_layout()
plt.show()
```

```python
# Use 'LabelEncoder' to change object type columns

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

df_new = df
df_new["Market Category"] = encoder.fit_transform(df["Market Category"])
df_new["Make"] = encoder.fit_transform(df["Make"])
df_new["Model"] = encoder.fit_transform(df["Model"])
df_new["Engine Fuel Type"] = encoder.fit_transform(df["Engine Fuel Type"])
df_new["Transmission Type"] = encoder.fit_transform(df["Transmission Type"])
df_new["Driven_Wheels"] = encoder.fit_transform(df["Driven_Wheels"])
df_new["Vehicle Size"] = encoder.fit_transform(df["Vehicle Size"])
df_new["Vehicle Style"] = encoder.fit_transform(df["Vehicle Style"])
```

```python
# Separate the feature columns and target column (MSRP)

features = df_new.drop(["MSRP"], axis=1)
target = df_new["MSRP"]
```

```python
# finding outliers in data

plt.figure(figsize=(20,20), facecolor = 'lightgreen')
graph = 1

for column in features:
    if graph < 16:   
        ax = plt.subplot(4,4,graph)
        sns.boxplot(data = features[column])
        plt.xlabel(column,fontsize=10)      
    graph+=1
plt.show()
```

```python
#calculation of IQR

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3-Q1
print(IQR)
```

```python
# removal of outliers

df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
#Task 1.A: Create a plot that shows the number of car models in each market category and their corresponding popularity scores.

# Group the data by market category and count the number of car models
market_counts = df.groupby('Market Category')['Model'].count()

# Group the data by market category and calculate the average popularity score
market_popularity = df.groupby('Market Category')['Popularity'].mean()

# Create a bar plot showing the number of car models and their popularity scores
plt.figure(figsize=(12, 6))
plt.bar(market_counts.index, market_counts.values, label='Number of Car Models')
plt.plot(market_popularity.index, market_popularity.values, marker='o', color='red', label='Popularity Score')
plt.xlabel('Market Category')
plt.ylabel('Count / Popularity Score')
plt.title('Number of Car Models and Popularity by Market Category')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
#Task 1.B: Create a combo chart that visualizes the relationship between market category and popularity.

# Create a scatter plot to visualize the relationship between market category and popularity
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Market Category', y='Popularity', alpha=0.7)
plt.xlabel('Market Category')
plt.ylabel('Popularity')
plt.title('Relationship between Market Category and Popularity')
plt.xticks(rotation=90)
plt.show()
```

```python
# Task 2:  Create a scatter chart that plots engine power on the x-axis and price on the y-axis. Add a trendline to the chart to visualize the relationship between these variables.

# Create a scatter plot with a trendline
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Engine HP', y='MSRP')
plt.xlabel('Engine Power')
plt.ylabel('Price')
plt.title('Relationship between Engine Power and Price')
plt.show()
```

```python
# Task 4.A: Create a chart for average price by manufacturer
chart_data = df.groupby('Make')['MSRP'].mean().reset_index()
```

```python
# Task 4.B: Create a bar chart for manufacturer and average price
plt.figure(figsize=(12, 6))
sns.barplot(data=chart_data, x='Make', y='MSRP')
plt.xlabel('Manufacturer')
plt.ylabel('Average Price')
plt.title('Average Price by Manufacturer')
plt.xticks(rotation=90)
plt.show()
```

```python
# Task 5.A: Create a scatter plot with number of cylinders vs. highway MPG
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Engine Cylinders', y='highway MPG')
plt.xlabel('Number of Cylinders')
plt.ylabel('Highway MPG')
plt.title('Number of Cylinders vs. Highway MPG')
plt.show()

# Task 5.B: Calculate the correlation coefficient between number of cylinders and highway MPG
correlation_coefficient = df['Engine Cylinders'].corr(df['highway MPG'])
print("Correlation Coefficient:", correlation_coefficient)
```
